### Pivot and merge multiple columns:

```python
def pivot_multi_columns(data, targets, index_col, column_col, aggfunc="mean", fill_value=None, dropna=False):
    # Precheck input data:
    assert isinstance(data, pd.DataFrame), TypeError("Input data must be pd.DataFrame.")
    
    # Pre-process input parameters:
    # index_col = [index_col] if not isinstance(index_col, list) else index_col
    
    # Make a placeholder for merging data:
    output = data[[index_col]]
    
    # Loop target columns to aggregate data:
    for ele in targets:
        try:
            tmp_pivot_df = pd.pivot_table(data, values=ele, index=index_col, \
                                          columns=column_col, aggfunc=aggfunc, \
                                          fill_value=fill_value, dropna=dropna).reset_index()
            tmp_pivot_df.columns = [ele + '_' + _col if _col != index_col else _col for _col in tmp_pivot_df.columns.tolist()]
            output = pd.merge(output, tmp_pivot_df, on=index_col, how="left")
        except Exception as e:
            print("Failed in pivot-merging data. Error: {}.".format(e))
            raise
            
    return output
```

#### 1. Merge non-transpose tables

```python
def merge_non_transpose_files(s3_path, s4_path, s3_files, s4_files, s3_data, s4_data):
    # placeholder for output:
    output = pd.DataFrame()
    files_need_transpose = []
    
    # First loop through s3 files:
    for _file in s3_files:
        # add path:
        # _file = s3_path + _file
        # add extension:
        # _file += extension # _file = _file + extension
        tmp_df = s3_data[_file]
        if tmp_df['MaskID'].nunique() < tmp_df.shape[0]:
            print("Skip {}, may need to transpose it.".format(_file))
            files_need_transpose.append(_file)
            continue
        # Process meta data:
        if tmp_df.shape[0] != 10251: # Hard-coded for now...
            print("Warning: For {}, the length of data is {}".format(_file, tmp_df.shape[0]))
        tmp_df["MaskID"] = tmp_df["MaskID"].astype(int)
        if output.shape[1] == 0:
            output = pd.concat([output, tmp_df])
        else:
            output = pd.merge(output, tmp_df, on="MaskID", how="left")
            
    # Second loop through s4 files:
    for _file in s4_files:
        # add path:
        # _file = s4_path + _file
        # same, add extension:
        # _file += extension
        tmp_df = s4_data[_file]
        if tmp_df['MaskID'].nunique() < tmp_df.shape[0]:
            print("Skip {}, may need to transpose it.".format(_file))
            files_need_transpose.append(_file)
            continue
        # Process meta data:
        if tmp_df.shape[0] != 10251:
            print("Warning: For {}, the length of data is {}".format(_file, tmp_df.shape[0]))
        tmp_df["MaskID"] = tmp_df["MaskID"].astype(int)
        if output.shape[1] == 0:
            output = pd.concat([output, tmp_df])
        else:
            output = pd.merge(output, tmp_df, on="MaskID", how="left")
            
    # output:
    return output, files_need_transpose
```

```python
merged_non_transpose_data, files_need_transpose = merge_non_transpose_files(s3_path=s3path, s3_files=s3files, s3_data=s3data, s4_path=s4path, s4_files=s4files, s4_data=s4data)
gc.collect()
```
